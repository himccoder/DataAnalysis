# Limitations of the dataset

Let's summarize our limitations of this dataset:


1. **Missing Data**:
   - **Reason**: Some columns have missing values due to incomplete data collection(e.g. MSRP).
   - This can skew the analysis and lead to biased conclusions if not handled properly.


2. **Geographic Limitations**:
   - **Reason**: Data is limited to Washington State.
   - This can lead to generalizations that do not account for regional differences in EV adoption, policies, or infrastructure.

3. **Categorical Data**:
   - **Reason**: Certain columns categorize vehicles by brand, model, or type without sufficient detail (e.g., hybrid vs. fully electric).
   - This can hinder analysis, such as understanding consumer preferences or performance data.

4. **Self-Reported Data**:
   - **Reason**: We aren't given any information on how the data is gathered!
   - Data set may be biased or inaccurate, affecting the reliability of conclusions drawn from this data. This in my opinion is the biggest limitation.

5. **Lack of Contextual Variables**:
   - **Reason**: The dataset lacks information on influencing factors such as charging infrastructure, government incentives, or fuel prices.
   - Without this context, it may be difficult to understand the factors driving EV adoption or the challenges faced.





# Investigating

Let's dive deeper and investigate the following questions:

1. Which EV manufacturers were the most popular in Washington?
2. How new are the EV models that are on the road in Washington?


```r
#GRAPH 1
# Group by Make and count the vehicles
make_distribution <- Electric_Vehicle %>%
  group_by(Make) %>%
  summarise(Vehicle_Count = n()) %>%
  arrange(desc(Vehicle_Count)) %>%
  top_n(10, Vehicle_Count)  # Show top 10 car makes

# Plot the distribution of EV manufacturers
ggplot(make_distribution, aes(x = reorder(Make, -Vehicle_Count), y = Vehicle_Count)) +
  geom_bar(stat = "identity", fill = "green") +
  labs(title ="Graph of top 10 Electric Vehicle Manufacturers", x = "Car Make", y = "Number of Vehicles") +
  theme_minimal()

#----------------------------------------

#GRAPH 2
# Group by Model Year to get the count of vehicles
model_year_distribution <- Electric_Vehicle %>%
  group_by(`Model Year`) %>%
  summarise(Vehicle_Count = n()) %>%
  arrange(`Model Year`)

# Plot the model year distribution
ggplot(model_year_distribution, aes(x = `Model Year`, y = Vehicle_Count)) +
  geom_line(color = "blue") +
  labs(title = "Vehicle Model Year Distribution", x = "Model Year", y = "Number of Vehicles") +
  theme_minimal()


```

Observing the first, we can identify that Tesla is dominating the EV market in Washington, above 75000 number of vehicles. The other car makers are at similiar levels all under 25000 cars, which include Chevrolet, Nissan, Ford , Kia etc.

The reason I chose this visualization is that a bar graph with the following scale used gives a good picture of Tesla's influence on the EV market in Washington State. We can extrapolate this visualization to get a sense of other US states. It gives a good picture of how the modern EV market is distributed and what are the major players.
